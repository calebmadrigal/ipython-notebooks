```python
import cmath

angle = numpy.linspace(0,2*math.pi, 32)

def generate_point(complex_num):
    return (complex_num.real, complex_num.imag)

def get_x_y(points):
    x = []
    y = []
    for p in points:
        x.append(p[0])
        y.append(p[1])
    return (x,y)
```
